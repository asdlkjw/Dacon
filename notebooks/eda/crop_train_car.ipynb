```python
from glob import glob
import cv2
import os
import pandas as pd
import numpy as np
```

```python
image_list = sorted(glob("/data/dacon_cars/data/train/*.png"))
label_list = sorted(glob("/data/dacon_cars/data/train/*.txt"))

print(len(image_list))
print(len(label_list))
```

```python
data_saved_path = "/data/dacon_cars/data/cars"
```

```python
for i in range(34):  # class 0~33
    os.makedirs(f"{data_saved_path}/{i}", exist_ok=True)
```

```python
for image_path, label_path in zip(image_list, label_list):
    image = cv2.imread(image_path)
    name = os.path.splitext(os.path.basename(image_path))[0]

    f_ann = open(label_path, "r")
    for i, line in enumerate(f_ann.readlines()):
        data = line.split()
        label = int(float(data[0]))
        pt1x = int(data[1])
        pt1y = int(data[2])
        pt3x = int(data[5])
        pt3y = int(data[6])

        crop_image = image[pt1y:pt3y, pt1x:pt3x]

        print(f"{data_saved_path}/{label}/{name}_{i}.png")

        # cv2.imwrite(f"{data_saved_path}/{label}/{name}_{i}.png", crop_image)
        np.save(f"{data_saved_path}/{label}/{name}_{i}.npy", crop_image)
```

```python
data_saved_path = "/data/dacon_cars/data/test_cars_image"
os.makedirs(data_saved_path, exist_ok=True)

df = pd.read_csv("/data/dacon_cars/data/bbox_submit_npy.csv")
df

i = 0
previous_name = ""
for _, row in df.iterrows():
    img = cv2.imread(f"/data/dacon_cars/data/test/{row['file_name']}")

    pt1x = int(np.round(row["point1_x"]))
    pt1y = int(np.round(row["point1_y"]))
    pt3x = int(np.round(row["point3_x"]))
    pt3y = int(np.round(row["point3_y"]))

    crop_image = img[pt1y:pt3y, pt1x:pt3x]

    name = os.path.splitext(row["file_name"])[0]
    if previous_name == name:
        i += 1
    else:
        previous_name = name
        i = 0

    print(f"{data_saved_path}/{name}_{i}.png")
    cv2.imwrite(f"{data_saved_path}/{name}_{i}.png", crop_image)
    # np.save(f"{data_saved_path}/{name}_{i}.png", crop_image)
    # break
```
